## The data
### **netflix_data.csv**
| Column | Description |
|--------|-------------|
| `show_id` | The ID of the show |
| `type` | Type of show |
| `title` | Title of the show |
| `director` | Director of the show |
| `cast` | Cast of the show |

| `date_added` | Date added to Netflix |
| `release_year` | Year of Netflix release |
| `duration` | Duration of the show in minutes |
| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Filter only movies
movies_df = netflix_df[netflix_df["type"] == "Movie"]
# Filter movies released in the 1990s
nineties_movies = movies_df[movies_df["release_year"].between(1990, 1999)]
# Get the most frequent movie duration in the 1990s
duration = nineties_movies["duration"].value_counts().idxmax()
# print the most frequent movie duration in the 1990s
print("Most frequent movie duration in the 1990s:", duration)
```

```python
# Filter movies released in the 1990s
nineties_movies = movies_df[movies_df["release_year"].between(1990, 1999)]

# Filter the 1990s movies to include only those that belong to the "Action" genre
action_movies_90s = nineties_movies[nineties_movies["genre"].str.contains("Action", na=False)]

# From the filtered action movies, select only those with a duration less than 90 minutes (short movies)
short_action_movies = action_movies_90s[action_movies_90s["duration"] < 90]

# Count the number of short action movies
short_movie_count = len(short_action_movies)

# Print number of short action movies in the 1990s
print("Number of short action movies in the 1990s:", short_movie_count)
```
